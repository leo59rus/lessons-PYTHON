```python
import pandas as pd
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data
```

```python
lstHuman = []
```

```python
lstRobot = []
```

```python

for i in data['whoAmI']:
  if i == 'human':
    lstHuman.append('1')
    lstRobot.append('0')
  else:
    lstRobot.append('1')
    lstHuman.append('0')
  
```

```python
data['human'] = lstHuman
data['robot'] = lstRobot
```
